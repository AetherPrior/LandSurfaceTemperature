```python
tot = []
```

```python
for i in os.listdir():
    tot.append(np.array(pd.read_csv(i, sep=" ", names=["x", "y", "Temp"])))
```

```python
tot = np.array(tot)
```

```python
a = tot.reshape(1568*2680, 3)
```

```python
ans = sorted(a, key= lambda p: p[0]**2 + p[1]**2)
```

```python
df_sort = pd.DataFrame(ans)
```

```python
df_sort. to_csv("sorted_final.csv")
```

```python
df_sort[1568*2678:1568*2680].describe().iloc[1][2]
```

```python
for i in range(0, 2680):
    if (df_sort[1568*i:1568*(i+1)].describe().iloc[-1][2] != 0):
        df_sort[1568*i:1568*(i+1)].to_csv("./pixel_timeseries/x_{}_y_{}.csv".format(df_sort[1568*i:1568*(i+1)].describe().iloc[-1][0], df_sort[1568*i:1568*(i+1)].describe().iloc[-1][1]))
```

```python
count = 0
for i in range(0, 2680):
    if (df_sort[1568*i:1568*(i+1)].describe().iloc[-1][2] != 0 and df_sort[1568*i:1568*(i+1)].describe().iloc[1][2] <= 5000):
        count += 1
```
